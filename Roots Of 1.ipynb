```julia
using Plots
gr();
n = 20;
anim = @animate for i in 1:n
    plot(sin,cos,-pi,pi,label="",aspect_ratio=1);
    scatter!(
        [ exp(2*pi*im*t/i) for t in 1:i],
        title="$i-th roots of 1",
        label=""
    )
end
gif(anim,"roots.gif",fps=1)
```

```julia
using Interact
@manipulate for i in 1:n
    plot(sin,cos,-pi,pi,label="",aspect_ratio=1);
    scatter!(
        [ exp(2*pi*im*t/i) for t in 1:i],
        title="$i-th roots of 1",
        label=""
    )
end
```
